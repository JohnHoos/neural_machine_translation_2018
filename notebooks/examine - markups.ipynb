## English - Chinese

```python
l = []
with open('vi-en/train.tok.en') as f:
    for line in f.readlines():
        l.append(line)
```

```python
w = []
for sent in l:
    
    w.extend(sent.split())
```

```python
i = 0
for word in w:
    if '&apos;' in word: i+=1
```

```python
a & b
a and b
```

```python
from collections import Counter
myCounter = Counter()
for word in w:
    if word.startswith('&') and not word.startswith('&apos;'):
        myCounter[word] += 1
myCounter
```

```python
for word in w:
    if word.startswith('1991'):
        myCounter[word] += 1
myCounter
```

## Vietnamese - En

```python
l_viet = []
with open('train.tok.vi') as f:
    for line in f.readlines():
        l_viet.append(line)
w_viet = []
for sent in l_viet:
    
    w_viet.extend(sent.split())
```

```python
w_viet
```

```python
myCounter_viet = Counter()
for word in w_viet:
    if word.startswith('1991') or word.startswith('&'):
        myCounter_viet[word] += 1
myCounter_viet
```
